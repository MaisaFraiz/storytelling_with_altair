### Exercise 2 - put it into words

The data for this exercise can be found here: https://www.storytellingwithdata.com/letspractice/downloads

```python
table = pd.read_excel(r"..\..\Data\6.2 EXERCISE.xlsx", usecols = [9, 10, 11, 12, 13], header = 4, skipfooter = 14)

table
```

```python
melted_table = pd.melt(table, id_vars = [2019], var_name = 'Metric', value_name = 'Value')
melted_table
```

```python
# Column names must be strings
melted_table.rename(columns = {2019:'2019'}, inplace = True)

selected_rows = melted_table[melted_table['Metric'].isin(['Accounts Worked', 'Dials Made'])]

alt.Chart(selected_rows).mark_bar().encode(
    y = alt.Y('Value'),
    x = alt.X('Metric'),
    color = alt.Color('Metric'),
    column = alt.Column('2019')
).configure_view(stroke = None)
```

```python
alt.Chart(selected_rows).mark_bar().encode(
    y = alt.Y('Value'),
    x = alt.X('2019'),
    color = alt.Color('Metric'),# scale = alt.Scale(range = ['#b4c6e4', '#4871b7'])),
    xOffset = 'Metric'
).configure_view(stroke = None)
```
